```python
from sqlalchemy import create_engine
import os
import pandas as pd

DB_SERVER = "(LocalDb)\\MSSQLLocalDB"
DB_NAME = 'Bank'

# Local Execution with Local Client
p = {
    'Server': DB_SERVER,
    'Database': DB_NAME
}

# Windows OS Driver selection for Local "VSCode/Shell" usage
if os.name == 'nt':
    driver = 'SQL Server Native Client 11.0'

# Linux OS Driver selection for Local "Analyticlab/AILab" usage
else:
    driver = 'ODBC Driver 17 for SQL Server'
 
# Local Engine Creation (User Name & Password are not required)
engine = create_engine(
    'mssql://' + p['Server'] + '/' + p['Database'] +'?driver=' + driver,
    echo=False,        
    fast_executemany=True)
print (engine)
```

```python
df1 = pd.read_excel('DB_TurkToSQL_[Cheque].xlsx', sheet_name='Sheet1')

display(df1)
```

```python
# Export the DataFrame to SQLite
df1.to_sql('Cheque', engine, schema='CHQ', index=False, if_exists='replace')
```

```python
# Verify the data is in the SQLite database
query = 'SELECT * FROM [Bank_deneme].[CHQ].[Cheque]'
result = pd.read_sql(query, engine)

# Print the result
print(result)
```
